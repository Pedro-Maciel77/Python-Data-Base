```python
import pandas as pd

# Carregando o arquivo CSV
df = pd.read_csv('StudentsPerformance.csv')


dados = df['reading score']
```

```python

# Dados de exemplo
dados = df['math score']

# Criando o histograma
plt.hist(dados, bins=5, color='skyblue', edgecolor='black')

# Adicionando títulos e rótulos
plt.title("Histograma de Frequência")
plt.xlabel("Valores")
plt.ylabel("Frequência")

# Exibindo o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Curva suavizada preenchida (KDE)
sns.kdeplot(dados, fill=True, color='red', edgecolor='black', alpha=0.3)

plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma com Curva Suavizada Preenchida')
plt.show()
```

```python
# Boxplot com outliers (padrão)
import matplotlib.pyplot as plt
import seaborn as sns

# Usando a mesma lista de dados do histograma
dados = df

plt.figure(figsize=(5, 5))
sns.boxplot(data=dados, showfliers=True)
plt.title('Boxplot com Outliers')
plt.ylabel('Valores')
plt.show()
```

```python
# Boxplot sem outliers
import matplotlib.pyplot as plt
import seaborn as sns

dados = df

plt.figure(figsize=(5, 5))
sns.boxplot(data=dados, showfliers=False)
plt.title('Boxplot sem Outliers')
plt.ylabel('Valores')
plt.show()
```
